## Задание 5
Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, ..., u_{n}\}$.

### Решение

Подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u+v\in L;$$

$$2)\ \alpha\cdot u \in L.$$

a) обозначим совокупность векторов $(0, a_1, b) + (a_0, 0, c) = (a_0, a_1, b + c)$ \
не совпадает с условием, т.к. только одна из первых двух координат может быть равна нулю, а в данном случае обе не равны нулю, значит такая совокупность не является линейным подпространством

b) Чтобы в этом убедиться, нужно проверить условия для минимум двух различных линейных комбинаций векторов

$\sum_{i=1}^{n}\lambda_{i}u_{i} + \sum_{j=1}^{n}\lambda_{j}u_{j} \in L$ Не выполняется для $\lambda_i \neq \lambda_j$

$\alpha \sum_{i=1}^{n}\lambda_{i}u_{i} \in L$ Выполняется

Следовательно не все векторы, являющиеся линейными комбинациями данных веторов формируют подпространство

## Задание 3
Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?

### Решение

Для Евклидова пространства должны выполняться следующие условия

1) $(x,y)=(y,x);$<br>
2) $(\lambda x, y)=\lambda(x, y);$<br>
3) $(x_{1}+x_{2},y)=(x_{1},y)+(x_{2},y);$<br>
4) $(x, x)\geq0$, причем $(x,x)=0\Leftrightarrow x=0.$

a) Если за скалярное произведение принять произведение длин векторов, то

$1. (x, y) = |x| + |y| = (y, x)$

$2. (\lambda x, y) = |\lambda||x||y| \neq \lambda(x, y) = \lambda(|x||y|)$

Второе условие не выполняется, значит пространство не является Евклидовым, если за скалярное произведение принять произведение длин векторов

b) Если за скалярное произведение принять обычное утроенное скалярное произведение векторов, то

1. $(x, y) = 3|x||y|cos\alpha = (y, x)$ \
2. $(\lambda x, y) = 3 \lambda |x||y|cos\alpha = \lambda(x, y)$ \
3. $(x_{1} + x_{2}, y)= 3 |x_1 + x_2||y|cos\alpha =(x_{1}, y)+(x_{2}, y)$
4. $(x, x)\geq0$, причем $(x,x)=0\Leftrightarrow x=0.$ выполняется ($3|x||x|cos\alpha = 0$, когда x = 0 или $cos\alpha = 0$, но в данном случае cos угла между двумя одинаковыми векторами всегда будет равен 1)

Выполняются все условия, следовательно пространство является Евклидовым

## Задание 4
Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

### Решение

В конечномерном евклидовом пространстве базис $\{e_{1},e_{2},...,e_{n}\}$ называется ортонормированным, если 

$(e_{i}, e_{j})=0$ $\forall$ $i\neq j$ и $(e_{i},e_{i})=1$ $\forall$ $i\in[1, n].$

а) 
$
\begin{cases}
(1, 0, 0) \cdot (0, 0, 1) = 0 \\
(1, 0, 0) \cdot (1, 0, 0) = 1 \\
(0, 0, 1) \cdot (0, 0, 1) = 1
\end{cases}\implies
$
образуют ортонормированный базис в линейном пространстве $R^3$

б)
$
\begin{cases}
(1/\sqrt{2},-1/\sqrt{2},0) \cdot (1/\sqrt{2},1/\sqrt{2},0) = 1/2 - 1/2 + 0 = 0\\
(1/\sqrt{2},1/\sqrt{2},0) \cdot (0,0,1) = 0\\
(1/\sqrt{2},-1/\sqrt{2},0) \cdot (0,0,1) = 0\\
(1/\sqrt{2},-1/\sqrt{2},0) \cdot (1/\sqrt{2},-1/\sqrt{2},0) = 1 \\
(1/\sqrt{2},1/\sqrt{2},0) \cdot (1/\sqrt{2},1/\sqrt{2},0) = 1\\
(0,0,1) \cdot (0,0,1) = 1
\end{cases}\implies
$ образуют ортонормированный базис в линейном пространстве $R^3$

в)
$
\begin{cases}
(1/2, -1/2, 0) \cdot (0, 1/2, 1/2) = -1/4\\
(0, 1/2, 1/2) \cdot (0,0,1) = 1/2 \\
(1/2, -1/2, 0) \cdot (0,0,1) = 0 \\
(1/2, -1/2, 0) \cdot (1/2, -1/2, 0) = 1 \\
(0, 1/2, 1/2) \cdot (0, 1/2, 1/2) = 1/2 \\
(0,0,1) \cdot (0,0,1) = 1
\end{cases}\implies
$ не образуют ортонормированный базис в линейном пространстве $R^3$

г)
$
\begin{cases}
(1, 0, 0) \cdot (0, 1, 0) = 0 \\
(0, 1, 0) \cdot (0, 0, 1) = 0 \\
(1, 0, 0) \cdot (0, 0, 1) = 0 \\
(1, 0, 0) \cdot (1, 0, 0) = 1 \\
(0, 1, 0) \cdot (0, 1, 0) = 1 \\
(0, 0, 1) \cdot (0, 0, 1) = 1
\end{cases}\implies
$ образуют ортонормированный базис в линейном пространстве $R^3$
