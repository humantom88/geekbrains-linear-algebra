## Практическое задание 1

__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

### Решение

$x = f_3(x) - f_2(x)$

$f_4(x)=f_3(x)-f_2(x)-f_1(x)$

$f_4(x) - f_3(x) + f_2(x) + f_1(x) = 0$

Векторы линейно независимы

$f_4(x)=x^2 + 2x + 1$

$f_4(x)=f_3(x) + 2f_2(x) + f_1(x) - 1$

$ f_3(x) + 2f_2(x) + f_1(x) - f_4(x) = 1$

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

   а) в базисе $1$, $x$, $x^{2}$;

   б) в базисе $x^{2}$, $x-1$, $1$.

Составим вектор из полинома: $[3, -2, 2]$

В базисе $1, x, x^2$: $$(3, 0, 0)\cdot(1, 0, 0) + (0, -2, 0)\cdot(0, x, 0) + (0, 0, 2)\cdot(0, 0, x^2) = (3, -2x, 2x^2)$$

В базисе $x^2, x - 1, 1$: $$(3, 0, 0)\cdot(x^2, 0, 0) + (0, -2, 0)\cdot(0, x - 1, 0) + (0, 0, 2)\cdot(0, 0, 1) = (3x^2, -2x + 2, 2)$$
